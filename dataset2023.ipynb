```python
import pandas as pd
from collections import defaultdict

dataframe = pd.read_excel('problemStatements.xlsx')
column_name = 'Description'

column_data = dataframe[column_name].tolist()
word_count = defaultdict(int)
conjunction_words = ['and', 'after', 'if', 'so', 'but', 'or', 'yet', 'for', 'nor','as','whether','also','because','since','though','that','once','how','nor','before','when','an','although','until','where','besides','&','!']

for description in column_data:
    words = description.split()
    for word in words:
        word_count[word] += 1

frequent_words = {word: count for word, count in word_count.items() if 5<= count <= 100}

sorted_frequent_words = dict(sorted(frequent_words.items(), key=lambda item: item[1]))

with open('frequent_words.txt', 'w') as file:
    for word, count in sorted_frequent_words.items():
        file.write(f"{word}: {count}\n")

```
